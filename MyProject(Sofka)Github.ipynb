```python
def check_answers(questions, questions2, questions3, questions4, questions5):
    score = 0
    random.shuffle(questions)
    for question in questions:
        answer = input(question.q).lower()
        if answer == question.answer:
            score += 1
            if score == 5:
                print("\n You pass to the second level \n")
        elif answer != question.answer:
            print("The game is over")
            break
    while(score==5):
        random.shuffle(questions2)
        for question in questions2:
            answer = input(question.q).lower()
            if answer == question.answer:
                score += 1   
                if score == 10:
                    print("\n You pass to the third level \n")
            else:
                print("The game is over") 
                break
        break
    
    while(score==10):
        random.shuffle(questions3)
        for question in questions3:
            answer = input(question.q).lower()
            if answer == question.answer:
                score += 1
                if score == 15:
                    print("\n You pass to the fourth level \n")
            else:
                print("The game is over")
                break
        break
    while(score==15):
        random.shuffle(questions4)
        for question in questions4:
            answer = input(question.q).lower()
            if answer == question.answer:
                score += 1
                if score == 20:
                    print("\n You pass to the fifth level (The Last One) \n")
            else:
                print("The game is over")
                break
        break
    while(score==20):
        random.shuffle(questions5)
        for question in questions5:
            answer = input(question.q).lower()
            if answer == question.answer:
                score += 1
                if score == 25:
                    print("\n Congrats! You win the game, you're the best \n")
            else:
                print("The game is over")
        break
    print("You got " + str(score) + "/" + str(len(questions)+len(questions2)+len(questions3)+len(questions4)+len(questions5)) + " correct")
     

    
#Defino una nueva función, inicializo la variable score = 0
#Utilice la función random.shuffle para elegir la pregunta al azar
#Iterar la matriz de la preguntas
#Compara la respuesta dada con la respuesta verdadera y si la respuesta es correcta, pasa a la siguiente pregunta; de lo contrario, imprime el juego ha terminado.
#Cuando se llega a 5 puntos pasa al nivel2, en 10 al nivel3, en 15 al nivel 4, en 20 al nivel 5 y por ultimo si saca todas las preguntas correctas, se genera un mensaje de felicitacion

```
